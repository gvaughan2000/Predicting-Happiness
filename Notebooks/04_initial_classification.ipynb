# 3 clusters

```python
# This data set is removing columns with more than 800 nans
# Then removes rows with Nans

X_train = pd.read_csv('../processed_data/3_clusters/x_train.csv')
y_train = pd.read_csv('../processed_data/3_clusters/y_train.csv').values.ravel()

X_test = pd.read_csv('../processed_data/3_clusters/X_test.csv')
y_test = pd.read_csv('../processed_data/3_clusters/y_test.csv').values.ravel()

display(X_train)
display(y_train)

LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
print('Linear Regression: ', round(LR.score(X_test,y_test), 4))

SVM = svm.SVC(decision_function_shape="ovo").fit(X_train, y_train)
print('SVM: ',round(SVM.score(X_test, y_test), 4))

RF = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0).fit(X_train, y_train)
print('Random Forest: ',round(RF.score(X_test, y_test), 4))

NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=1).fit(X_train, y_train)
print('NN: ',round(NN.score(X_test, y_test), 4))
```

# Comparing

## Nans removed

```python
X_train = pd.read_csv('../processed_data/nans_removed/x_train.csv')
y_train = pd.read_csv('../processed_data/nans_removed/y_train.csv').values.ravel()

X_test = pd.read_csv('../processed_data/nans_removed/X_test.csv')
y_test = pd.read_csv('../processed_data/nans_removed/y_test.csv').values.ravel()

models_df = make_models(X_train, y_train)
removed_nans_comparison = compare_models(models_df, X_train, y_train, X_test, y_test)
display(removed_nans_comparison)
```

```python
for i in [5, 10, 50, 100, 250, 500, 1000]:
    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train, y_train)
    y_pred = KNN.predict(X_test)
    print(metrics.accuracy_score(y_test, y_pred))
```

```python
comparison_df = pd.DataFrame(data = {'Name':removed_nans_comparison['Name'], 
                                     'Nans removed':removed_nans_comparison['Accuracy']})
```

# With Nans

```python
X_train = pd.read_csv('../processed_data/with_nans/x_train.csv')
y_train = pd.read_csv('../processed_data/with_nans/y_train.csv').values.ravel()

X_test = pd.read_csv('../processed_data/with_nans/X_test.csv')
y_test = pd.read_csv('../processed_data/with_nans/y_test.csv').values.ravel()

XGB = XGBClassifier(objective='binary:logistic')
LGB = LGBMClassifier().fit(X_train,y_train)
CatBoost = CatBoostClassifier(verbose=0, n_estimators=100).fit(X_train,y_train)

mini_models_dict = {'ExtremeGradientBoost': XGB,
                  'Light Gradient Boost': LGB,
                  'Cat Boost':CatBoost}

with_nans_comparison = compare_models(mini_models_dict, X_train, y_train, X_test, y_test)
display(with_nans_comparison)

#comparison_df['With Nans'] = comparison_df.merge(with_nans_comparison.set_index('Name')['Accuracy'])
#display(comparison_df)
```

```python
with_nans_dict = with_nans_comparison.set_index('Name')['Accuracy'].to_dict()
comparison_df['With Nans']= comparison_df['Name'].map(with_nans_dict)
```

### Imputation

```python
for method in ['most_frequent', 'mean', 'median', 'knn']:
    
    x_test_imputed = pd.read_excel(f'../processed_data/imputed/{method}/x_test.xlsx')
    x_train_imputed = pd.read_excel(f'../processed_data/imputed/{method}/x_train.xlsx')
    y_train = pd.read_excel(f'../processed_data/imputed/{method}/y_train.xlsx')
    y_test = pd.read_excel(f'../processed_data/imputed/{method}/y_test.xlsx')
    
    # creating the model
    all_models = make_models(x_train_imputed, y_train)
    model_scores_df = compare_models(all_models, x_train_imputed, y_train, x_test_imputed, y_test)
    display(model_scores_df)
    
    comparison_df[f'Imputation: {method}'] = model_scores_df['Accuracy']
    print(method, 'done')
```
